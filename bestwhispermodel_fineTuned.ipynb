```python
#   FINAL FINE-TUNED SER PIPELINE (Whisper-small + MLP-Mixer)

!pip install -q transformers[torch] librosa soundfile tqdm scikit-learn matplotlib seaborn

import os, random, time
import numpy as np
from tqdm.auto import tqdm
import librosa
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight

# -----------------------------
# SETTINGS
# -----------------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(SEED)

DATASET_PATH = "/kaggle/input/cremad/AudioWAV"
MODEL_PATH = "/kaggle/working/ser_finetuned.pt"
PLOT_DIR = "/kaggle/working/plots_finetuned"
os.makedirs(PLOT_DIR, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

emotion_map = {
    "ANG": "angry",
    "DIS": "disgust",
    "FEA": "fear",
    "HAP": "happy",
    "NEU": "neutral",
    "SAD": "sad"
}

# -----------------------------
# LOAD WHISPER ENCODER
# -----------------------------
from transformers import WhisperProcessor, WhisperModel
processor = WhisperProcessor.from_pretrained("openai/whisper-small")
whisper = WhisperModel.from_pretrained("openai/whisper-small").to(device)
whisper.eval()
print("Loaded Whisper-small encoder.")

# -----------------------------
# EMBEDDING EXTRACTION
# -----------------------------
def extract_whisper_embedding(path, target_sr=16000):
    audio, sr = librosa.load(path, sr=target_sr)

    # Normalize loudness
    if np.max(np.abs(audio)) > 0:
        audio = audio / np.max(np.abs(audio))

    # Remove leading & trailing silence
    audio, _ = librosa.effects.trim(audio)

    inputs = processor(audio, sampling_rate=target_sr, return_tensors="pt")
    feats = inputs.input_features.to(device)

    with torch.no_grad():
        out = whisper.encoder(feats)
        emb = out.last_hidden_state.mean(dim=1).cpu().numpy().squeeze()
    return emb

# Extract all embeddings 
files = sorted([f for f in os.listdir(DATASET_PATH) if f.endswith(".wav")])
X_list, y_list = [], []
for fname in tqdm(files, desc="Extracting Whisper Embeddings"):
    emo_code = fname.split("_")[2]
    emo = emotion_map.get(emo_code)
    if emo:
        path = os.path.join(DATASET_PATH, fname)
        emb = extract_whisper_embedding(path)
        X_list.append(emb)
        y_list.append(emo)

X = np.stack(X_list)
y = np.array(y_list)

print("Embeddings shape:", X.shape)
print("Labels:", np.unique(y))

# -----------------------------
# ENCODE LABELS & SPLIT
# -----------------------------
le = LabelEncoder()
y_enc = le.fit_transform(y)
classes = list(le.classes_)
print("Classes:", classes)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.20, stratify=y_enc, random_state=SEED
)

# Compute class weights
class_weights = compute_class_weight('balanced', classes=np.unique(y_enc), y=y_enc)
class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)

# Torch loaders
BATCH = 64
train_ds = TensorDataset(torch.tensor(X_train).float(), torch.tensor(y_train).long())
test_ds = TensorDataset(torch.tensor(X_test).float(), torch.tensor(y_test).long())

train_loader = DataLoader(train_ds, batch_size=BATCH, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=BATCH, shuffle=False)

# MLP-MIXER INSPIRED CLASSIFIER
class MixerSER(nn.Module):
    def __init__(self, dim, num_classes=6):
        super().__init__()
        self.block1 = nn.Sequential(
            nn.LayerNorm(dim),
            nn.Linear(dim, 1024),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(1024, dim),
        )
        self.block2 = nn.Sequential(
            nn.LayerNorm(dim),
            nn.Linear(dim, 512),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(512, dim),
        )
        self.final = nn.Sequential(
            nn.LayerNorm(dim),
            nn.Linear(dim, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = x + self.block1(x)
        x = x + self.block2(x)
        return self.final(x)

model = MixerSER(dim=X.shape[1], num_classes=len(classes)).to(device)
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-2)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=50)

print(model)

# -----------------------------
# TRAINING LOOP
# -----------------------------
EPOCHS = 50
train_losses, val_losses, train_accs, val_accs = [], [], [], []
best_val_loss = float("inf")
patience, wait = 8, 0

for epoch in range(1, EPOCHS+1):
    model.train()
    total, correct, running_loss = 0, 0, 0
    for xb, yb in tqdm(train_loader, desc=f"Train Epoch {epoch}", leave=False):
        xb, yb = xb.to(device), yb.to(device)

        optimizer.zero_grad()
        out = model(xb)
        loss = criterion(out, yb)
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        running_loss += loss.item() * xb.size(0)
        correct += (out.argmax(1) == yb).sum().item()
        total += xb.size(0)

    train_losses.append(running_loss / total)
    train_accs.append(correct / total)

    # Validation
    model.eval()
    v_loss, v_correct, v_total = 0, 0, 0
    with torch.no_grad():
        for xb, yb in test_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            loss = criterion(out, yb)
            v_loss += loss.item() * xb.size(0)
            v_correct += (out.argmax(1) == yb).sum().item()
            v_total += xb.size(0)

    val_losses.append(v_loss / v_total)
    val_accs.append(v_correct / v_total)

    print(f"Epoch {epoch}/{EPOCHS} | "
          f"Train Acc {train_accs[-1]:.4f} | Val Acc {val_accs[-1]:.4f} | "
          f"Train Loss {train_losses[-1]:.4f} | Val Loss {val_losses[-1]:.4f}")

    scheduler.step()

    # Early stopping
    if val_losses[-1] < best_val_loss:
        best_val_loss = val_losses[-1]
        wait = 0
        torch.save({"model": model.state_dict(), "classes": classes}, MODEL_PATH)
    else:
        wait += 1
        if wait >= patience:
            print("Early stopping triggered.")
            break

print("Best model saved to:", MODEL_PATH)

# -----------------------------
# PLOTS
# -----------------------------
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_accs); plt.plot(val_accs)
plt.title("Accuracy"); plt.legend(["Train","Val"])
plt.subplot(1,2,2)
plt.plot(train_losses); plt.plot(val_losses)
plt.title("Loss"); plt.legend(["Train","Val"])
plt.savefig(os.path.join(PLOT_DIR, "curves.png"))
plt.show()

# -----------------------------
# EVALUATION
# -----------------------------
def predict(loader):
    model.eval()
    preds, trues = [], []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            out = model(xb)
            preds.extend(out.argmax(1).cpu().numpy())
            trues.extend(yb.numpy())
    return np.array(trues), np.array(preds)

train_t, train_p = predict(train_loader)
test_t, test_p = predict(test_loader)

print("\nTRAIN REPORT:")
print(classification_report(train_t, train_p, target_names=classes, digits=4))

print("\nTEST REPORT:")
print(classification_report(test_t, test_p, target_names=classes, digits=4))

# Confusion matrix
cm_test = confusion_matrix(test_t, test_p)
plt.figure(figsize=(7,6))
sns.heatmap(cm_test, annot=True, cmap="Blues", fmt="d", xticklabels=classes, yticklabels=classes)
plt.title("Test Confusion Matrix")
plt.savefig(os.path.join(PLOT_DIR, "cm_test.png"))
plt.show()
```

```python
# ========= TRAIN CONFUSION MATRIX (Standalone Cell) ========= #

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Ensure model is in eval mode
model.eval()

# Function to get predictions from a loader
def get_train_preds(loader):
    preds_all = []
    targets_all = []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            logits = model(xb)
            preds = logits.argmax(dim=1).cpu().numpy()
            preds_all.extend(preds)
            targets_all.extend(yb.numpy())
    return np.array(preds_all), np.array(targets_all)

# Get predictions for training set
train_preds_only, train_targets_only = get_train_preds(train_loader)

# Compute confusion matrix
cm_train_only = confusion_matrix(train_targets_only, train_preds_only)

# Plot
plt.figure(figsize=(7,6))
sns.heatmap(
    cm_train_only, 
    annot=True, 
    fmt='d', 
    cmap='Blues',
    xticklabels=classes,
    yticklabels=classes
)
plt.title("Confusion Matrix - TRAIN ")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```
