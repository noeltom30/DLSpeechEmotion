```python
# (Whisper-small + MLP)

!pip install -q transformers[torch] librosa soundfile tqdm scikit-learn matplotlib seaborn

# 1) Imports & settings
import os, random, time
import numpy as np
from tqdm.auto import tqdm
import librosa
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(SEED)

# Paths (adjust if needed)
DATASET_PATH = "/kaggle/input/cremad/AudioWAV"
CACHE_PATH = "/kaggle/working/cremad_whisper_embeddings.npz"
MODEL_PATH = "/kaggle/working/cremad_ser_model.pt"
PLOT_DIR = "/kaggle/working/plots"
os.makedirs(PLOT_DIR, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# 2) CREMA-D mapping
emotion_map = {
    "ANG": "angry",
    "DIS": "disgust",
    "FEA": "fear",
    "HAP": "happy",
    "NEU": "neutral",
    "SAD": "sad"
}

# 3) Load Whisper encoder (processor + model)
from transformers import WhisperProcessor, WhisperModel
print("Loading Whisper-small (encoder)... This may take a moment.")
processor = WhisperProcessor.from_pretrained("openai/whisper-small")
whisper = WhisperModel.from_pretrained("openai/whisper-small").to(device)
whisper.eval()

# 4) Embedding extraction helper (mean-pooled encoder output)
def extract_whisper_embedding(path, sr_target=16000):
    audio, sr = librosa.load(path, sr=sr_target)
    if audio.size == 0:
        raise RuntimeError(f"Empty audio: {path}")
    inputs = processor(audio, sampling_rate=sr_target, return_tensors="pt")
    input_features = inputs.input_features.to(device)  # shape: (1, seq_len, feat_dim)
    with torch.no_grad():
        enc = whisper.encoder(input_features)  # enc.last_hidden_state: (1, time, dim)
        emb = enc.last_hidden_state.mean(dim=1).cpu().numpy().squeeze()
    return emb

# 5) Load or compute embeddings (with progress bar)
if os.path.exists(CACHE_PATH):
    print("Loading cached embeddings:", CACHE_PATH)
    cache = np.load(CACHE_PATH, allow_pickle=True)
    X = cache["X"]
    y = cache["y"]
    print("Loaded embeddings:", X.shape)
else:
    print("No cache found. Extracting embeddings (this will take time)...")
    files = sorted([f for f in os.listdir(DATASET_PATH) if f.lower().endswith(".wav") or f.lower().endswith(".mp3")])
    X_list = []
    y_list = []
    bad = []
    for fname in tqdm(files, desc="Files"):
        try:
            parts = fname.split("_")
            emo_code = parts[2] if len(parts) > 2 else ""
            emo = emotion_map.get(emo_code.upper())
            if emo is None:
                continue
            path = os.path.join(DATASET_PATH, fname)
            emb = extract_whisper_embedding(path)
            X_list.append(emb)
            y_list.append(emo)
        except Exception as e:
            bad.append((fname, str(e)))
    if len(X_list) == 0:
        raise RuntimeError("No embeddings extracted. Check DATASET_PATH and filename format.")
    X = np.stack(X_list, axis=0)
    y = np.array(y_list)
    np.savez_compressed(CACHE_PATH, X=X, y=y)
    print(f"Saved {X.shape[0]} embeddings to cache. Bad files: {len(bad)}")
    if len(bad)>0:
        print("Examples of bad files:", bad[:5])

# Basic sanity checks
print("Embedding shape:", X.shape)
if np.isnan(X).any():
    raise RuntimeError("Found NaNs in embeddings. Aborting.")
print("Unique labels:", np.unique(y))

# 6) Encode labels and split
le = LabelEncoder()
y_enc = le.fit_transform(y)
classes = list(le.classes_)
print("Classes (label encoder):", classes)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.20, stratify=y_enc, random_state=SEED
)
print("Train / Test sizes:", X_train.shape[0], X_test.shape[0])

# 7) Torch datasets & loaders
BATCH_SIZE = 32
train_ds = TensorDataset(torch.tensor(X_train, dtype=torch.float32),
                         torch.tensor(y_train, dtype=torch.long))
test_ds  = TensorDataset(torch.tensor(X_test,  dtype=torch.float32),
                         torch.tensor(y_test,  dtype=torch.long))
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, drop_last=False)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False)

# 8) Model (MLP) — expects input shape (batch, feature_dim)
class MLPClassifier(nn.Module):
    def __init__(self, input_dim, hidden=1024, n_classes=6, drop=0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Dropout(drop),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Dropout(drop),
            nn.Linear(hidden//2, n_classes)
        )
    def forward(self, x):
        return self.net(x)

model = MLPClassifier(input_dim=X.shape[1], hidden=1024, n_classes=len(classes)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
print(model)

# 9) Training loop with clear logs and simple early stopping option
EPOCHS = 30
PATIENCE = 6   # set to None or an int; if None, no early stopping
best_val_loss = float("inf")
no_improve = 0

train_losses, val_losses, train_accs, val_accs = [], [], [], []

for epoch in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    t0 = time.time()
    pbar = tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} Train", leave=False)
    for xb, yb in pbar:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        logits = model(xb)
        loss = criterion(logits, yb)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * xb.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == yb).sum().item()
        total += xb.size(0)
        pbar.set_postfix({'batch_loss': running_loss/total, 'batch_acc': correct/total})

    epoch_train_loss = running_loss / total
    epoch_train_acc  = correct / total
    train_losses.append(epoch_train_loss)
    train_accs.append(epoch_train_acc)

    # Validation
    model.eval()
    v_loss = 0.0; v_correct = 0; v_total = 0
    with torch.no_grad():
        for xb, yb in tqdm(test_loader, desc=f"Epoch {epoch}/{EPOCHS} Val", leave=False):
            xb, yb = xb.to(device), yb.to(device)
            logits = model(xb)
            loss = criterion(logits, yb)
            v_loss += loss.item() * xb.size(0)
            preds = logits.argmax(dim=1)
            v_correct += (preds == yb).sum().item()
            v_total += xb.size(0)
    epoch_val_loss = v_loss / v_total
    epoch_val_acc  = v_correct / v_total
    val_losses.append(epoch_val_loss)
    val_accs.append(epoch_val_acc)

    t1 = time.time()
    print(f"Epoch {epoch}/{EPOCHS} — time {t1-t0:.1f}s | Train loss {epoch_train_loss:.4f} acc {epoch_train_acc:.4f} | Val loss {epoch_val_loss:.4f} acc {epoch_val_acc:.4f}")

    # simple early stopping
    if PATIENCE is not None:
        if epoch_val_loss < best_val_loss - 1e-6:
            best_val_loss = epoch_val_loss
            no_improve = 0
            # save best model checkpoint
            torch.save({'model_state': model.state_dict(), 'classes': classes}, MODEL_PATH)
        else:
            no_improve += 1
        if no_improve >= PATIENCE:
            print(f"No improvement for {PATIENCE} epochs. Early stopping.")
            break

# if early stopping disabled, save final model
if PATIENCE is None:
    torch.save({'model_state': model.state_dict(), 'classes': classes}, MODEL_PATH)
print("Saved model checkpoint to", MODEL_PATH)

# 10) Plot accuracy & loss curves and save
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_accs, label="train_acc")
plt.plot(val_accs, label="val_acc")
plt.xlabel("epoch"); plt.title("Accuracy"); plt.legend()

plt.subplot(1,2,2)
plt.plot(train_losses, label="train_loss")
plt.plot(val_losses, label="val_loss")
plt.xlabel("epoch"); plt.title("Loss"); plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(PLOT_DIR, "acc_loss.png"))
plt.show()

# 11) Predictions function
def get_preds(loader):
    model.eval()
    preds_all = []
    targets_all = []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            logits = model(xb)
            preds_all.extend(logits.argmax(dim=1).cpu().numpy())
            targets_all.extend(yb.numpy())
    return np.array(preds_all), np.array(targets_all)

train_preds, train_targets = get_preds(train_loader)
test_preds,  test_targets  = get_preds(test_loader)

# 12) Classification reports
print("\n=== TRAIN CLASSIFICATION REPORT ===")
print(classification_report(train_targets, train_preds, target_names=classes, digits=4))

print("\n=== TEST CLASSIFICATION REPORT ===")
print(classification_report(test_targets, test_preds, target_names=classes, digits=4))

# 13) Confusion matrices (your requested style)
cm_train = confusion_matrix(train_targets, train_preds)
plt.figure(figsize=(7,6))
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title("Confusion Matrix - Train")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout()
plt.savefig(os.path.join(PLOT_DIR, "cm_train.png"))
plt.show()

cm_test = confusion_matrix(test_targets, test_preds)
plt.figure(figsize=(7,6))
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title("Confusion Matrix - Test")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout()
plt.savefig(os.path.join(PLOT_DIR, "cm_test.png"))
plt.show()

print("\nOutputs (plots & model) saved to:", PLOT_DIR, "and", MODEL_PATH)
```
